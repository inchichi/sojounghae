# Key Results

## Portable Schema Comparison
- `billing2 = monthly_billing + total_billing`
- `billing3 = billing2 + payment_card`

## Main Findings
- `billing2` was the best portable schema.
- `payment_card` hurt Cell2Cell -> IBM transfer in raw form.
- `rank normalization` and `CORAL` improved Cell2Cell -> IBM transfer for `billing2`.
- IBM -> Cell2Cell remained close to chance, so transfer is still asymmetric.

## Compact Table
| Schema | Method | Cell2Cell -> IBM AUC | IBM -> Cell2Cell AUC | Domain AUC |
|---|---|---:|---:|---:|
| billing2 | raw | 0.6061 | 0.5000 | 0.9940 |
| billing2 | rank | 0.6396 | 0.5243 | 0.9540 |
| billing2 | CORAL | 0.6837 | 0.5256 | 0.8902 |
| billing3 | raw | 0.3327 | 0.4931 | 0.9975 |
| billing3 | rank | 0.5771 | 0.5254 | 0.9695 |
| billing3 | CORAL | 0.5820 | 0.5139 | 1.0000 |

## Interpretation
The portable schema should be fixed around `billing2`. `payment_card` adds domain-specific noise, while rank normalization / CORAL reduce but do not eliminate domain mismatch.
