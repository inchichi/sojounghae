# Portable Billing + Credit Extension

## Fixed Baseline
- Portable baseline: `billing2 + CORAL`
- `billing2 = monthly_billing + total_billing`

## Credit Extension
- `billing2_credit_score = billing2 + credit_score`
- `billing2_credit_family = billing2_credit_score + credit_access`
- `credit_score` is ordinalized from `CreditRating` in Cell2Cell and `PaymentMethod` trust order in IBM Telco.
- `credit_access` is `HasCreditCard` in Cell2Cell and automatic-payment adoption in IBM Telco.

## Key Result
| Schema | Method | Cell2Cell -> IBM AUC | IBM -> Cell2Cell AUC | Domain AUC |
|---|---|---:|---:|---:|
| billing2 | CORAL | 0.6837 | 0.5256 | 0.8902 |
| billing2_credit_score | CORAL | 0.4796 | 0.5267 | 0.9993 |
| billing2_credit_family | CORAL | 0.5108 | 0.5116 | 1.0000 |

## Conclusion
- `billing2 + CORAL` remains the best portable baseline.
- Adding the credit axis increases domain separability sharply and does not improve cross-domain transfer.
- The credit family is therefore not adopted in the final portable schema.
