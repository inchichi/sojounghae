# IBM -> Cell2Cell Reverse Transfer

IBM Telco in-domain 모델을 Cell2Cell 데이터에 역방향으로 적용한 실험 결과를 정리한다.

- IBM in-domain holdout: IBM Telco 내부 검증 성능
- Reverse transfer: IBM Telco 모델을 Cell2Cell에 적용한 성능


```python
import pandas as pd

rows = [
    ['IBM in-domain holdout', 0.8388, 0.7473, 0.5162, 0.7674, 0.6172, 0.625, 0.6227, '-'],
    ['IBM -> Cell2Cell reverse transfer', 0.4821, 0.6426, 0.2626, 0.1328, 0.1764, 0.05, 0.4175, 0.5179],
]
df = pd.DataFrame(rows, columns=['Setting', 'ROC-AUC', 'Accuracy', 'Precision', 'Recall', 'F1', 'Best Threshold', 'Best F1', 'Inverted AUC'])
print(df.to_markdown(index=False))
```

## 해석

- IBM in-domain 성능은 높지만 Cell2Cell 역전이는 ROC-AUC 0.5 아래로 떨어진다.
- 이전 Cell2Cell -> IBM 전이보다 덜 무너지긴 하지만, 실용적인 transfer 모델로 보기는 어렵다.
- 두 통신사 데이터는 문제 정의는 비슷해도 feature 의미와 경계가 다르다는 뜻이다.
