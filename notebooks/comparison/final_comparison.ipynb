# 최종 비교 노트북

지금까지의 실험을 한 번에 비교하는 요약 노트북입니다.
모델 가족 기준의 베이스라인과 XGBoost 단계별 변화를 같이 봅니다.


```python
import pandas as pd
from pathlib import Path

FIG_DIR = Path('/home/ms990728/소정해/figures')

family_df = pd.DataFrame([
    {'Model': 'Logistic Regression', 'ROC-AUC': 0.6084, 'Accuracy': 0.5848, 'Precision': 0.3625, 'Recall': 0.5809, 'F1': 0.4464},
    {'Model': 'Random Forest', 'ROC-AUC': 0.6676, 'Accuracy': 0.7200, 'Precision': 0.5737, 'Recall': 0.1098, 'F1': 0.1843},
    {'Model': 'XGBoost', 'ROC-AUC': 0.6820, 'Accuracy': 0.6309, 'Precision': 0.4087, 'Recall': 0.6292, 'F1': 0.4955},
])

stage_df = pd.DataFrame([
    {'Stage': 'Baseline XGBoost', 'ROC-AUC': 0.6820, 'Accuracy': 0.6309, 'Precision': 0.4087, 'Recall': 0.6292, 'F1': 0.4955, 'Threshold': 0.500},
    {'Stage': 'Engineered XGBoost', 'ROC-AUC': 0.6794, 'Accuracy': 0.6305, 'Precision': 0.4073, 'Recall': 0.6203, 'F1': 0.4917, 'Threshold': 0.500},
    {'Stage': 'Selected XGBoost', 'ROC-AUC': 0.6741, 'Accuracy': 0.6248, 'Precision': 0.4037, 'Recall': 0.6332, 'F1': 0.4931, 'Threshold': 0.500},
    {'Stage': 'Tuned XGBoost', 'ROC-AUC': 0.6804874753395962, 'Accuracy': 0.5897159647404505, 'Precision': 0.3877587758775878, 'Recall': 0.7321549966009517, 'F1': 0.5070024714605155, 'Threshold': 0.465},
])
stage_df['ΔROC-AUC vs Baseline'] = stage_df['ROC-AUC'] - stage_df.loc[0, 'ROC-AUC']
stage_df['ΔF1 vs Baseline'] = stage_df['F1'] - stage_df.loc[0, 'F1']

print('Model family baseline')
print(family_df.to_string(index=False))
print()
print('XGBoost stage comparison')
print(stage_df.to_string(index=False))

# Bar chart for stage comparison
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-whitegrid')
fig, axes = plt.subplots(1, 2, figsize=(12, 4.8), sharey=True)
colors = ['#334155', '#64748b', '#94a3b8', '#1d4ed8']
axes[0].barh(stage_df['Stage'], stage_df['ROC-AUC'], color=colors, edgecolor='none')
axes[0].set_title('ROC-AUC by Stage')
axes[0].set_xlim(0.64, 0.70)
axes[0].spines[['top', 'right']].set_visible(False)
for y, v in enumerate(stage_df['ROC-AUC']):
    axes[0].text(v + 0.0008, y, f'{v:.4f}', va='center', fontsize=9)
axes[1].barh(stage_df['Stage'], stage_df['F1'], color=colors, edgecolor='none')
axes[1].set_title('F1 by Stage')
axes[1].set_xlim(0.45, 0.52)
axes[1].spines[['top', 'right']].set_visible(False)
for y, v in enumerate(stage_df['F1']):
    axes[1].text(v + 0.0008, y, f'{v:.4f}', va='center', fontsize=9)
fig.suptitle('XGBoost Experiment Comparison', y=1.02, fontsize=15)
fig.tight_layout()
fig.savefig(FIG_DIR / 'xgb_stage_comparison.png', bbox_inches='tight', facecolor='white')
plt.show()
```
